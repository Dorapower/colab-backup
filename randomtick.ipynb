```python
import math
```

```python
class ChunkPos:
  x,z = 0.,0.

  def __init__(self,x,z):
    self.x=x%16
    self.z=z%16
  def __repr__(self):
    return '({x:.2f}, {z:.2f})'.format(x=self.x,z=self.z)
```

```python
def MinMaxLimitation(x:float, radius:float=128, shift:float=0):
  delta = math.sqrt(radius**2-shift**2)
  return x-delta, x+delta
```

```python
# excluding the point exactly on the edge
def MinMaxChunkCoordinate(x:float, radius:float=128, shift:float=0):
  chunk_size=16
  minl, maxl = MinMaxLimitation(x,radius,shift) # float limitation
  min =  ( minl-chunk_size/2)//chunk_size+1
  max =-((-maxl+chunk_size/2)//chunk_size+1)
  return int(min), int(max)
```

```python
# calculate number of random ticked chunk with player chunk position
def chunk_count(pos:ChunkPos):
  chunk_size = 16
  north, south = MinMaxChunkCoordinate(pos.z)
  cnt = 0
  for chunkz in range(north, south+1):
    shift = chunk_size*(chunkz+1/2)-pos.z
    west, east = MinMaxChunkCoordinate(pos.x,shift=shift)
    len = east-west+1
    cnt+=len
  return cnt
```

```python
list(map(chunk_count,[ChunkPos(0,0),ChunkPos(0.1,0.1),ChunkPos(7.9,7.9),ChunkPos(8,8)]))
```

```python
# plot the result
import plotly.graph_objects as go
import numpy as np
N=640
X=np.linspace(0,16,N,endpoint=False)
Y=np.linspace(0,16,N,endpoint=False)
Z=np.zeros(shape=(N,N))
for x in range(N):
  for y in range(N):
    Z[x,y]=chunk_count(ChunkPos(X[x],Y[y]))
fig = go.Figure(data=go.Surface(z=Z,x=X,y=Y))
```
